# 3. Volume vs Value Analysis
In this section, we use **Pandas** to compare the physical volume of goods moved (Units Sold) against the financial value generated (Revenue). We also calculate the Average Revenue Per Unit to evaluate pricing realization.

````python
import pandas as pd

# 1. Data Preparation
df = pd.read_excel('mdgsalesdata.xlsx')
target_products = ['DOVE NUTRITIVE', 'SUNSILK EXPERT']
df_filtered = df[df['Product'].isin(target_products)]

# 2. Comparing Units Sold (Volume) with Revenue (Value)
val_df = df_filtered.groupby('Product')[['Units_Sold', 'Revenue']].sum().reset_index()

# 3. Calculating average revenue per unit
val_df['Avg_Revenue_Per_Unit'] = round(val_df['Revenue'] / val_df['Units_Sold'], 2)

print(val_df)
````

### Results Output:
```text
          Product  Units_Sold  Revenue  Avg_Revenue_Per_Unit
0  DOVE NUTRITIVE       16358   479760                 29.33
1  SUNSILK EXPERT       14171   419490                 29.60
```

### Business Insights & Conclusions:
* **Volume Leadership:** **DOVE NUTRITIVE** leads in both volume (16,358 units) and total value, confirming its strong market penetration compared to SUNSILK EXPERT.
* **Price Realization & Trade Spend:** The baseline recommended retail price (RRP) for these products is **30 MAD**. However, the realized Average Revenue Per Unit is **29.33 MAD** for Dove and **29.60 MAD** for Sunsilk. 
* **Margin Leakage:** The difference of 0.67 MAD (Dove) and 0.40 MAD (Sunsilk) per unit represents trade discounts and promotional markdown. This clearly reveals the commercial strategy: Dove achieved its higher sales volume by utilizing slightly heavier discounting (0.67 MAD/unit) compared to Sunsilk.
