# Part 1: First Tertile Analysis (Days 1 - 10)
Exploring the initial trends for DOVE NUTRITIVE and SUNSILK EXPERT using Pandas. This analysis focuses on the first 10 days of the month to identify early momentum and anomalies.

```python
import pandas as pd

# Simulating the extracted data for the first tertile based on the dataframe
data_t1 = {
    'Final Date': ['2026-07-01']*2 + ['2026-07-02']*2 + ['2026-07-03']*2 + ['2026-07-04']*2 + ['2026-07-05']*2 + ['2026-07-06']*2 + ['2026-07-07']*2 + ['2026-07-08']*2 + ['2026-07-09']*2 + ['2026-07-10']*2,
    'Product': ['DOVE NUTRITIVE', 'SUNSILK EXPERT'] * 10,
    'Revenue': [12930, 15930, 18120, 11190, 14460, 13920, 15660, 10650, 16860, 11310, 16950, 15990, 21210, 13230, 17940, 16230, 25170, 18540, 2550, 18390]
}

df_t1 = pd.DataFrame(data_t1)

# Grouping and Analysis using Pandas (.agg function for multiple metrics)
summary_t1 = df_t1.groupby('Product')['Revenue'].agg(['sum', 'mean', 'max', 'min']).reset_index()

print("--- First Tertile Summary ---")
print(summary_t1)
```

*   **Revenue Leadership:** **DOVE NUTRITIVE** leads the first tertile with a total revenue of **161,850 MAD**, compared to **SUNSILK EXPERT** which generated **145,380 MAD**.
*   **Peak Performance:** Dove showed a strong upward momentum leading up to `2026-07-09`, where it hit its peak revenue of **25,170 MAD**.
*   **Critical Anomaly Detected:** On `2026-07-10`, Dove's sales plummeted drastically to just **2,550 MAD** (its lowest point). Given that Sunsilk maintained a strong performance on the same day (18,390 MAD), this sharp drop is likely an isolated issue such as a sudden stock-out or a data recording error, rather than a general drop in market demand.
