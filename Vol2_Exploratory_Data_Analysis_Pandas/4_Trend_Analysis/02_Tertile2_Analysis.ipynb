# Part 2: Second Tertile Analysis (Days 11 - 20)
Exploring the mid-month trends for DOVE NUTRITIVE and SUNSILK EXPERT using Pandas. This section highlights revenue stabilization and captures any new anomalies following the first tertile.

```python
import pandas as pd

# Simulating the extracted data for the second tertile
data_t2 = {
    'Final Date': ['2026-07-11']*2 + ['2026-07-12']*2 + ['2026-07-13']*2 + ['2026-07-14']*2 + ['2026-07-15']*2 + ['2026-07-16']*2 + ['2026-07-17']*2 + ['2026-07-18']*2 + ['2026-07-19']*2 + ['2026-07-20']*2,
    'Product': ['DOVE NUTRITIVE', 'SUNSILK EXPERT'] * 10,
    'Revenue': [17070, 3630, 13650, 15120, 15210, 14790, 12390, 16170, 11610, 13470, 14430, 17850, 19740, 11910, 14580, 15030, 14670, 11460, 14070, 14370]
}

df_t2 = pd.DataFrame(data_t2)

# Grouping and Analysis using Pandas (.agg function for multiple metrics)
summary_t2 = df_t2.groupby('Product')['Revenue'].agg(['sum', 'mean', 'max', 'min']).reset_index()

print("--- Second Tertile Summary ---")
print(summary_t2)
```

### Business Insights & Conclusion (Mid-Month Trends)

*   **Revenue Stabilization & Leadership:** After the sharp drop at the end of the first tertile, **DOVE NUTRITIVE** successfully recovered and maintained a stable revenue stream. It remained the top performer, generating a total revenue of **147,420 MAD** over these 10 days.
*   **Anomaly Shift (Sunsilk's Turn):** Interestingly, the anomaly pattern shifted brands. On `2026-07-11`, **SUNSILK EXPERT** experienced a severe crash, generating only **3,630 MAD** (its lowest point of the month so far). This mirror-like anomaly, compared to Dove's previous drop, strongly suggests a systemic issue—such as a localized supply chain disruption or inventory rotation—affecting one brand at a time, rather than a decline in general consumer demand.
*   **Mid-Month Consistency:** Excluding the anomaly on the 11th, both products demonstrated stable daily revenues fluctuating comfortably between 11K and 19K, indicating predictable consumer behavior during the middle of the month.
