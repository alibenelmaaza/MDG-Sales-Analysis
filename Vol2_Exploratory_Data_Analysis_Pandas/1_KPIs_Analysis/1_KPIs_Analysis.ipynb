# 1. KPIs Analysis (Achievement Rate)
In this section, we calculate the total revenue generated by each product and compare it against the predefined targets to find the **Achievement Rate %**.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Data Preparation
df = pd.read_excel('sales_moderntrade_n1.xlsx')
target_products = ['DOVE NUTRITIVE', 'SUNSILK EXPERT']
df_filtered = df[df['Product'].isin(target_products)]
```

```python
# Grouping data by product to calculate total revenue and target
kpi_df = df_filtered.groupby('Product')[['Revenue', 'Target']].sum().reset_index()

# Calculating Achievement Rate Percentage
kpi_df['Achievement_Rate_%'] = round((kpi_df['Revenue'] / kpi_df['Target']) * 100, 2)

print(kpi_df)
```

```python
# Data Visualization
plt.figure(figsize=(8, 5))
sns.barplot(data=kpi_df, x='Product', y='Achievement_Rate_%', palette='Blues_d')
plt.axhline(100, color='red', linestyle='--', label='Target (100%)')
plt.title('Achievement Rate %', fontsize=14)
plt.ylabel('Percentage (%)')
plt.xlabel('Product')
plt.legend()
plt.tight_layout()
plt.show()
```

### Business Insights: KPIs & Achievement Rate
* **Overall Performance:** Both products fell short of their optimal 100% revenue targets. This indicates a potential need to review the initial sales forecasting models or investigate external market headwinds.
* **Top Performer:** **DOVE NUTRITIVE** stands out as the stronger product, generating higher total revenue and leading with an achievement rate of **87.95%**.
* **Area for Improvement:** **SUNSILK EXPERT** slightly lags behind with an **85.02%** target completion, signaling that this specific product might require targeted promotional efforts or strategic marketing adjustments to boost its market penetration.
