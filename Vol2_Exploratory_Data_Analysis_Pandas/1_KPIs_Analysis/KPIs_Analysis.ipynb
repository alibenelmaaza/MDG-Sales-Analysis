# 1. KPIs Analysis (Achievement Rate)
In this section, we use **Pandas** to aggregate the total revenue generated by each product and compare it against the predefined targets to calculate the **Achievement Rate %**.

````python
import pandas as pd

# 1. Data Preparation
df = pd.read_excel('mdgsalesdata.xlsx')
target_products = ['DOVE NUTRITIVE', 'SUNSILK EXPERT']
df_filtered = df[df['Product'].isin(target_products)]

# 2. Grouping data by product to calculate total revenue and target
kpi_df = df_filtered.groupby('Product')[['Revenue', 'Target']].sum().reset_index()

# 3. Calculating Achievement Rate Percentage
kpi_df['Achievement_Rate_%'] = round((kpi_df['Revenue'] / kpi_df['Target']) * 100, 2)

print(kpi_df)
````

### Results Output:
```text
          Product  Revenue  Target  Achievement_Rate_%
0  DOVE NUTRITIVE   479760  545519               87.95
1  SUNSILK EXPERT   419490  493383               85.02
```

### Business Insights & Conclusions:
* **Overall Performance:** Both products fell short of their 100% revenue targets. This indicates a potential need to review the initial sales forecasting models or investigate external market headwinds.
* **Top Performer:** **DOVE NUTRITIVE** stands out as the stronger product, generating higher total revenue (479,760 MAD) and leading with an achievement rate of **87.95%**.
* **Area for Improvement:** **SUNSILK EXPERT** lags behind with an **85.02%** target completion, signaling that this specific product might require targeted promotional efforts or strategic marketing adjustments to boost its market penetration.
