# Sales Performance & Market Analysis
**Target:** Comparing 'DOVE NUTRITIVE' and 'SUNSILK EXPERT'

#### Methodology
1. **Data Preparation:** Load data and filter target products.
2. **KPIs & Achievement Rate:** Calculate total revenue vs target.
3. **Variance Analysis:** Measure the absolute difference between actual sales and goals.
4. **Volume vs Value:** Compare units sold to revenue generated to extract pricing insights.
5. **Trend Analysis:** Track daily sales performance.

```python
import pandas as pd

# 1. Data Preparation
df = pd.read_excel('mdgsalesdata.xlsx')
df['Final Date'] = pd.to_datetime(df['Final Date'])

target_products = ['DOVE NUTRITIVE', 'SUNSILK EXPERT']
df_filtered = df[df['Product'].isin(target_products)]

print("Data loaded and filtered successfully!")
```

### 1. KPIs Analysis (Achievement Rate)

```python
# Grouping data by product to calculate total revenue and target
kpi_df = df_filtered.groupby('Product')[['Revenue', 'Target']].sum().reset_index()

# Calculating Achievement Rate Percentage
kpi_df['Achievement_Rate_%'] = round((kpi_df['Revenue'] / kpi_df['Target']) * 100, 2)

print(kpi_df)
```

### 2. Variance Analysis

```python
# Calculating the actual difference between Revenue and Target
var_df = df_filtered.groupby('Product')[['Revenue', 'Target']].sum().reset_index()
var_df['Variance'] = var_df['Revenue'] - var_df['Target']

print(var_df)
```

### 3. Volume vs Value Analysis

```python
# Comparing Units Sold (Volume) with Revenue (Value)
val_df = df_filtered.groupby('Product')[['Units_Sold', 'Revenue']].sum().reset_index()

# Calculating average revenue per unit
val_df['Avg_Revenue_Per_Unit'] = round(val_df['Revenue'] / val_df['Units_Sold'], 2)

print(val_df)
```

### 4. Trend Analysis

```python
# Tracking daily revenue for each product
trend_df = df_filtered.groupby(['Final Date', 'Product'])['Revenue'].sum().reset_index()

print(trend_df.head(10))
```

### Business Conclusions & Insights

* **Market Performance (KPIs & Variance):** By analyzing the achievement rate and variance, we can easily identify which product is driving the most profit and which one is lagging behind its set targets, allowing for quick strategic pivots.
* **Pricing Strategy (Volume vs Value):** A high volume of units sold doesn't necessarily mean higher profitability. The 'Avg_Revenue_Per_Unit' metric provides a clear view of the actual value each product brings to the business, which is crucial for evaluating discounting and promotional strategies.
* **Sales Dynamics (Trend Analysis):** Daily tracking helps in identifying peak sales days, enabling better inventory management and more targeted marketing campaigns.
