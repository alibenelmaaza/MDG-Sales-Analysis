# 2. Variance Analysis
In this section, we use **Pandas** to calculate the absolute financial difference (Variance) between actual sales and target expectations for each product.

````python
import pandas as pd

# 1. Data Preparation
df = pd.read_excel('sales_moderntrade_n1.xlsx')
target_products = ['DOVE NUTRITIVE', 'SUNSILK EXPERT']
df_filtered = df[df['Product'].isin(target_products)]

# 2. Grouping data by product to calculate total revenue and target
var_df = df_filtered.groupby('Product')[['Revenue', 'Target']].sum().reset_index()

# 3. Calculating Variance (Actual Revenue - Target)
var_df['Variance'] = var_df['Revenue'] - var_df['Target']

print(var_df)
````

### Results Output:
```text
          Product  Revenue  Target  Variance
0  DOVE NUTRITIVE   479760  545519    -65759
1  SUNSILK EXPERT   419490  493383    -73893
```

### Business Insights & Conclusions:
* **Financial Shortfall:** Both products display a negative variance, translating to a direct financial shortfall. The combined missed revenue is substantial and requires immediate review of the current sales pipeline.
* **The Sunsilk Paradox:** Despite having a significantly lower revenue target than DOVE NUTRITIVE, **SUNSILK EXPERT** generated a larger financial deficit (a variance of -73,893 MAD). This indicates underperformance in the market and suggests that Sunsilk should be the primary focus for corrective marketing campaigns or trade promotions.
